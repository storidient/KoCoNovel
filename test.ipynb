```python
import json
from pathlib import Path
from collections import defaultdict
```

```python
def read_jsonl(path):  
    with open(path, "r", encoding="utf-8") as f:
        data = [json.loads(line) for line in f]
    return data
```

```python
for x in Path('./data/jsonl/').glob('*'):
    coref_path = x / 'coref.jsonl'
    text_path = x / 'text.jsonl'
    coref = read_jsonl(coref_path)
    text_dict = {doc['doc_id'] : doc['text'] for doc in read_jsonl(text_path)}
    
    for doc in coref:
        speakers = sum(doc['speakers'], [])
        if len(speakers) == 0:
            continue
        
        reader_clusters = doc['reader_overlapped']
        clusters = doc['omniscent_separate']
        print(reader_clusters)
        print(clusters)
```
